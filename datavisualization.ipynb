```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("final.csv")
df = pd.DataFrame(data)
```

```python
avg_pts = df.groupby('Tm')['PTS'].mean()

# Bar plot
plt.figure(figsize=(10, 6))
avg_pts.plot(kind='bar', color='green')
plt.title('Average Points per Game by Team')
plt.xlabel('Team')
plt.ylabel('Average Points per Game')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=5, color='purple', alpha=0.7)
plt.title('Distribution of Player Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df.boxplot(column='PTS', by='Pos', patch_artist=True, showmeans=True)
plt.title('Points Scored by Position')
plt.xlabel('Position')
plt.ylabel('Points Scored')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['MP'], df['PTS'], color='blue', alpha=0.5)
plt.title('Points vs Minutes Played')
plt.xlabel('Minutes Played')
plt.ylabel('Points Scored')
plt.grid(True)
plt.show()
```

```python
average_share_by_year = df.groupby('Year')['Share'].mean()

# Sort by descending average win shares and select top 150
top_150_average_share = average_share_by_year.sort_values(ascending=False).head(100)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(top_150_average_share.index, top_150_average_share.values, marker='o', linestyle='-', color='b', label='Average Win Shares')
plt.title('Top 150 Average Win Shares per Year')
plt.xlabel('Year')
plt.ylabel('Average Win Shares')
plt.xticks(top_150_average_share.index, rotation=45)  # Rotate x-axis labels by 45 degrees
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
